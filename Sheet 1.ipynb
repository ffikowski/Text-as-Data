```python
with open('reviews3.txt', encoding='utf16') as f:
    reviews3 = f.readlines()
```

```python
reviews = []

reviews.extend([reviews1[0],reviews1[1]])
```

```python
reviews.extend([reviews2[0],reviews2[1]])
```

```python
reviews.extend([reviews3[0],reviews3[1]])
```

```python
print(len(reviews))
```

## Task 2:

1. Application of the elementary tokenizations steps:

```python
preprocessed_reviews = []

for k in np.arange(len(reviews)):
    text= ""
    for i in np.arange(len(reviews[k])):
        if reviews[k][i].isalpha() or reviews[k][i].isspace():
            text += reviews[k][i].lower()
    text_split = text.split()
    preprocessed_reviews.append(text_split)
print(preprocessed_reviews)
```

2. Count the most frequent words

```python
counter = collections.Counter([])
for i in preprocessed_reviews:
    review_counter = collections.Counter(i)
    counter += review_counter
```

```python
counter_filtered = dict((k, v) for k, v in counter.items() if v >= 5)
```

```python
first5pairs = {k: counter_filtered[k] for k in list(counter_filtered)[:5]}
print(first5pairs)
```

## Task 3:

1. Application of a stemming method:

```python
stemmer = SnowballStemmer("english")
```

```python
print(stemmer.stem(preprocessed_reviews[0][0]))
```

```python
stemmed_reviews = []

for i in preprocessed_reviews:
    stemmed_review = []
    for k in i:
        stemmed_token = stemmer.stem(k)
        stemmed_review.append(stemmed_token)
    stemmed_reviews.append(stemmed_review)
print(stemmed_reviews)
```

```python
counter_stemmed = collections.Counter([])
for i in stemmed_reviews:
    review_counter = collections.Counter(i)
    counter_stemmed += review_counter
```

```python
counter_stemmed_filtered = dict((k, v) for k, v in counter_stemmed.items() if v >= 5)
```

```python
print(counter_filtered)
```

2. Application of a lemmatization method:

```python
lemmatizer = WordNetLemmatizer()
```

```python
lammatized_reviews = []

for i in preprocessed_reviews:
    lammatized_review = []
    for k in i:
        lemmatized_token = lemmatizer.lemmatize(k)
        lammatized_review.append(lemmatized_token)
    lammatized_reviews.append(lammatized_review)
print(lammatized_reviews)
```

```python
for i in np.arange(len(preprocessed_reviews)):
    for k in np.arange(len(preprocessed_reviews[i])):
        print(preprocessed_reviews[i][k],stemmed_reviews[i][k], lammatized_reviews[i][k])
```

```python
# I would prefer lemmatization, since the words look more natural compared to stemming.
```

## Task 4:

```python
stop_words = list(set(stopwords.words('english')))
```

```python
print(stop_words)
```

```python
preprocessed_stop_words = []

for k in np.arange(len(stop_words)):
    text= ""
    for i in np.arange(len(stop_words[k])):
        if stop_words[k][i].isalpha() or stop_words[k][i].isspace():
            text += stop_words[k][i].lower()
    preprocessed_stop_words.append(text)
print(preprocessed_stop_words)
```

```python
filtered_reviews = []

for i in np.arange(len(lammatized_reviews)):
    filtered_review = []
    for k in lammatized_reviews[i]:
        if k not in preprocessed_stop_words:
            filtered_review.append(k)
    filtered_reviews.append(filtered_review)
```

```python
print(filtered_reviews)
```

```python
counter_filtered_stemmed = collections.Counter([])
for i in filtered_reviews:
    review_counter = collections.Counter(i)
    counter_filtered_stemmed += review_counter
```

```python
counter_stemmed_filtered = dict((k, v) for k, v in counter_filtered_stemmed.items() if v >= 5)
```
