## Task 3:

```python
corpus_joined = []


for i in np.arange(len(corpus)):
    joined_doc = ' '.join(corpus[i])
    print(type(joined_doc))
    corpus_joined += [joined_doc]
    print(len(corpus_joined))
```

```python
regex_list = ['(Page \S \d{1,3}( \n){0,})([a-zA-Z0-9_ ]{1,})(-( ){0,}J.K. Rowling )',' [A-Z| ]{2,}\n \n','\S( \n){3,}','\n']

for i in np.arange(len(corpus_joined)):
    print("Before:",len(corpus_joined[i]))
    for k in np.arange(len(regex_list)):
        corpus_joined[i] = re.sub(regex_list[k], '', corpus_joined[i], count=0, flags=0)
    print("After:",len(corpus_joined[i]))    
```

## Task 4:

```python
preprocessed_books = []

for k in np.arange(len(corpus_joined)):
    print("Before:",len(corpus_joined[k]))
    text= ""
    for i in np.arange(len(corpus_joined[k])):
        if corpus_joined[k][i].isalpha() or corpus_joined[k][i].isspace():
            text += corpus_joined[k][i].lower()
    text_split = text.split()
    preprocessed_books.append(text_split)
    print("After:",len(text_split))    
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lammatized_books = []

for i in preprocessed_books:
    print("Before:",len(i))
    lammatized_book = []
    for k in i:
        lemmatized_token = lemmatizer.lemmatize(k)
        lammatized_book.append(lemmatized_token)
    lammatized_books.append(lammatized_book)
    print("After:",len(lammatized_book))
```

```python
stop_words = list(set(stopwords.words('english')))
```

```python
preprocessed_stop_words = []

for k in np.arange(len(stop_words)):
    text= ""
    for i in np.arange(len(stop_words[k])):
        if stop_words[k][i].isalpha() or stop_words[k][i].isspace():
            text += stop_words[k][i].lower()
    text = lemmatizer.lemmatize(text)
    preprocessed_stop_words.append(text)
print(len(preprocessed_stop_words))
```

```python
filtered_books = []

for i in np.arange(len(lammatized_books)):
    print("Before:",len(lammatized_books[i]))
    filtered_book = []
    for k in lammatized_books[i]:
        if k not in preprocessed_stop_words:
            filtered_book.append(k)
    filtered_books.append(filtered_book)
    print("After:",len(filtered_book))
```

```python
print(len(filtered_books))
```

```python
for i in np.arange(len(filtered_books)):
    value = ' '.join(filtered_books[i])
    filtered_books[i] = value
```

## Task 5:

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(filtered_books)
```

```python
def get_top_tfidf_words(matrix, vocabulary, top_n=7):
    dense = matrix.toarray()
    results = []
    for row in dense:
        row_indices = np.argsort(row)[::-1][:top_n]
        row_words = [vocabulary[index] for index in row_indices]
        results.append(row_words)
    return results
```

```python
feature_names = vectorizer.get_feature_names()
top_words = get_top_tfidf_words(X,feature_names)
```
